```python
#Prime numbers

def checkPrime(num):
    if num>1:
        for i in range(2,num):
            if(num%i)==0:
                print(num," is not a prime number")
                break
            else:
                print(num," is a prime number")
                break
    else:
        print(num," is not a prime number")

checkPrime(10)
```

```python
lst=list(range(2500))
print(lst)
```

```python
lst1=filter(checkPrime,lst)
print(list(lst1))
```
